**Question 1**
Write a python code to determine how many 3's are in the number 3^1000.

```python
import numpy as np

n = 3 ** 1000
n = str(n)

count = 0
for i in range(len(n)):
    if n[i] == '3':
        count = count + 1

print(count)
```

**Question 2**
Create a numpy array x with values from 0 (inclusive) to 10 (inclusive). Successive x values should be separated by 0.1. Create an array y of the same length as x, where y = sin(x) + r and the elements of r are random real numbers from -0.1 (inclusive) to 0.1 (exclusive). Plot a graph of y vs x. 

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 101)

rnum = np.random.default_rng()
r = []
for i in range(101):
    r = np.append(r, rnum.uniform(-0.1, 0.1))

y = np.sin(x) + r

plt.close()
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = sin(x) + r")
plt.show()
```

**Question 3**
Download the file test1data.txt from moodle. The file consists of a single column of numerical data. Half the numbers are positive and half are negative. Separate the data into two arrays, a "plus" array containing all the positive numbers, and a "minus" array containing all the negative numbers. Add the "plus" and "minus" arrays together, element by element, and print the result. 

```python
import numpy as np 

a = np.loadtxt("test1data.txt")
plus = []
minus = []
for i in range(len(a)):
    if a[i] < 0:
        minus = np.append(minus, a[i])
    else: # there are no 0s so we don't need to account for that 
        plus = np.append(plus, a[i])

combined = plus + minus # this should work if they have the same number of elements
print(combined)
```

**Question 4**
Create a function definition for

    f(n) = n/2, if n is even 
           n/3, if n is not even and is divisible by 3
           5n + 1, if n is not even and is not divisible by 3

In your main code, start with an initial choice for n, compute f(n), and iterate. That is, use f(n) as the new value of n, evaluate the function again, and repeat. Have your code print out each integer in the sequence, and stop iterating if the integer reaches 1 or 6 or 3. (The "5n + 1" conjecture says taht for any initial n, the sequence will eventually reach the cycle 1, 6, or 3). 

```python

def f(n):
    if n % 2 == 0: 
        return n / 2
    else: 
        if n % 3 == 0:
            return n / 3
        else: 
            return 5 * n + 1

n = 432
while n != 1 and n != 6 and n != 3:
    print(n)
    n = f(n)

print(n)
```
