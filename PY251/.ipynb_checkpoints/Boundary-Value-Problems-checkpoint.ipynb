```python
import math
import matplotlib.pyplot as plt

# Parameters
mu_g_over_Tx = 5.0
xA, yA = 0.0, 0.0
xB, yB = 1.0, 2.0
N = 101
sweeps = 5000

dx = (xB - xA) / (N - 1)

# Trial solution: straight line y = 2x
y = []
for i in range(N):
    y.append(2.0 * (i * dx))

# store y[51] every 1000 sweeps
mid_vals = {} # dictionary: 1000 --> key (k), value --> y[51]

xs = []
for i in range(N):
    value = i * dx
    xs.append(value)

for s in range(1, sweeps + 1):

    for i in range(1, N - 1):
        dy_dx = (y[i+1] - y[i-1]) / (2 * dx)
        RHS = math.sqrt(1 + dy_dx**2)
        y[i] = 0.5 * (y[i+1] + y[i-1]) - 0.5 * mu_g_over_Tx * dx**2 * RHS # Applying 20.15 to each of the nodes

    # Plot every 1000 sweeps
    if s % 1000 == 0:
        mid_vals[s] = y[51]
        plt.plot(xs, y, label=f"{s} sweeps")

print("Midpoint values:")
for k, v in mid_vals.items():
    print(f"Sweep {k}: y[51] = {v}")

print("\nDifferences:")
print("1000 to 2000:", mid_vals[2000] - mid_vals[1000])
print("2000 to 3000:", mid_vals[3000] - mid_vals[2000])
print("3000 to 4000:", mid_vals[4000] - mid_vals[3000])
print("4000 to 5000:", mid_vals[5000] - mid_vals[4000])


plt.xlabel("x")
plt.ylabel("y")
plt.title("Exercise 20.3")
plt.legend()
plt.show()
```

**Exercise 20.4a**

ANSWER: THe GaussSeidel relaxes faster because it has a larger change from the initial condition. 

```python
import math
import numpy as np

# Problem Parameters
mu_g_over_Tx = 5.0
N = 101
sweeps = 5000
dx = 1.0 / (N - 1)
mid = N // 2 # (51)

# Trial solution: straight line y = 2x
def initial_guess():
    y = []
    for i in range(N):
        y.append(2.0 * (i * dx))
    return y

# Initial guess
initial = initial_guess()[mid]

# Right side of differential equation
def RHS(y, i):
    dy_dx = (y[i+1] - y[i-1]) / (2 * dx)
    return math.sqrt(1 + dy_dx**2)

# Gauss-Seidel Method (uses newly updated values)
y_gs = initial_guess()

for s in range(sweeps):
    for i in range(1, N - 1):
        rhs = RHS(y_gs, i)
        y_gs[i] = 0.5 * (y_gs[i+1] + y_gs[i-1]) - 0.5 * mu_g_over_Tx * dx**2 * rhs

# Jacobi Method (uses only old values each sweep)
y_jac = initial_guess()

for s in range(sweeps):
    old = np.copy(y_jac)
    for i in range(1, N - 1):
        rhs = RHS(old, i)       
        y_jac[i] = 0.5 * (old[i+1] + old[i-1]) - 0.5 * mu_g_over_Tx * dx**2 * rhs

# Report midpoint results
print("Midpoint values after 5000 sweeps:")
gs_diff = abs(initial - y_gs[mid])
jac_diff = abs(initial - y_jac[mid])
print(f"GaussSeidel : y[51] = {y_gs[mid]}, Change from initial: {gs_diff}")
print(f"Jacobi : y[51] = {y_jac[mid]}, Change from initial: {jac_diff}")
```

**Exercise 20.5b**

ANSWER: 
- With well-chosen omega values (for this problem, this seems to be around 1.2-1.6), SOR converges much faster than standard Gauss-Seidel (which is omega = 1). You can get similar accuracy in about 2000 to 3000 sweeps with SOR that takes 5000 sweeps with Gauss-Seidel.
- If omega gets too large, the values start becoming nonsensical (it seems to be about omega = 2, in this case). 

```python
import math
import numpy as np

# Problem Parameters
mu_g_over_Tx = 5.0
N = 101
sweeps = 5000
dx = 1.0 / (N - 1)
mid = N // 2  # index 51

# Initial solution: straight line y = 2x
def initial_guess():
    y = []
    for i in range(N):
        y.append(2.0 * (i * dx))
    return y

# Right side of differential equation
def RHS(y, i):
    dy_dx = (y[i+1] - y[i-1]) / (2 * dx)
    return math.sqrt(1 + dy_dx**2)

# Gauss-Seidel Method
def gauss_seidel(sweeps):
    y = initial_guess()
    for s in range(sweeps):
        for i in range(1, N - 1):
            rhs = RHS(y, i)
            y[i] = 0.5 * (y[i+1] + y[i-1]) - 0.5 * mu_g_over_Tx * dx**2 * rhs
    return y

# SOR Method
def sor(sweeps, omega):
    y = initial_guess()
    for s in range(sweeps):
        for i in range(1, N - 1):
            y_old = y[i]
            rhs = RHS(y, i)
            y_new = 0.5 * (y[i+1] + y[i-1]) - 0.5 * mu_g_over_Tx * dx**2 * rhs # From Jacobi method 
            # SOR, Eq 20.16
            y[i] = omega * y_new + (1 - omega) * y_old
    return y

# Omega values test
omega_vals = np.linspace(0.1, 2, 20)

print("Comparison of Gauss-Seidel and SOR methods")

# Run Gauss-Seidel
y_gs = gauss_seidel(sweeps)

print(f"\nSOR with different  values:")

for s in range(1000, 5001, 1000):
    print(f"\nSweeps = {s}:")
    for omega in omega_vals:
        y_sor = sor(s, omega)
        diff_from_gs = abs(y_sor[mid] - y_gs[mid])
        print(f" = {omega:.1f}: y[51] = {y_sor[mid]:.6f}, "
              f"diff from GS = {diff_from_gs:.6f}")
```

```python
import math
import numpy as np

# Problem Parameters
mu_g_over_Tx = 5.0
N = 101
dx = 1.0 / (N - 1)
mid = N // 2  # index 51

# Initial solution: straight line y = 2x
def initial_guess():
    y = []
    for i in range(N):
        y.append(2.0 * (i * dx))
    return y

# Right side of differential equation
def RHS(y, i):
    dy_dx = (y[i+1] - y[i-1]) / (2 * dx)
    return math.sqrt(1 + dy_dx**2)

# Eq 20.17
def compute_residual(y, i):
    second_deriv = (y[i+1] - 2*y[i] + y[i-1]) / (dx**2)
    dy_dx = (y[i+1] - y[i-1]) / (2 * dx)
    rhs_term = mu_g_over_Tx * math.sqrt(1 + dy_dx**2)
    residual = second_deriv - rhs_term
    return residual

# Gauss-Seidel Method with residual monitoring
def gauss_seidel_residual(epsilon, max_sweeps=50000):

    # Top part - from original function 
    y = initial_guess()
    for s in range(1, max_sweeps + 1):
        for i in range(1, N - 1):
            rhs = RHS(y, i)
            y[i] = 0.5 * (y[i+1] + y[i-1]) - 0.5 * mu_g_over_Tx * dx**2 * rhs

        # Print
        if s % 1000 == 0:
            print(f"{s:<10} {max_residual:<15.6f} {y[mid]:<15.6f}")
        
        # Compute maximum residual
        max_residual = 0.0
        for i in range(1, N - 1):
            r = abs(compute_residual(y, i))
            if r > max_residual:
                max_residual = r
        
        # Check for convergence
        if max_residual < epsilon:
            # Print if not already printed so that final value show up 
            if s % 100 != 0:  
                print(f"{s:<10} {max_residual:<15.6f} {y[mid]:<15.6f}")
            return s, y
    
    # If it didn't converge in 50000 sweeps for some reason (adjust 50000 if it doesn't converge)
    return max_sweeps, y


# Gauss-Seidel with epsilon = 0.001
print(" = 0.001")
print("-----------------")
sweeps_001, y_001 = gauss_seidel_residual(epsilon=0.001)

print("")

# Gauss-Seidel with epsilon = 0.0001
print(" = 0.0001")
print("-----------------")
sweeps_0001, y_0001 = gauss_seidel_residual(epsilon=0.0001)

print("")

# Summary
print(f"epsilon = 0.001:  {sweeps_001} sweeps required")
print(f"epsilon = 0.0001: {sweeps_0001} sweeps required")
print(f"\ny[51] with epsilon = 0.001:  {y_001[mid]:.6f}")
print(f"y[51] with epsilon = 0.0001: {y_0001[mid]:.6f}")
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Problem setup
xA, yA = 0.0, 0.0
xB, yB = 1.0, 2.0
N = 101
sweeps = 5000
dx = (xB - xA) / (N - 1)

# Initial solution: straight line y = 2x
def initial_guess():
    y = []
    for i in range(N):
        y.append(2.0 * (i * dx))
    return y

# Right side of differential equation
def RHS(y, i):
    dy_dx = (y[i+1] - y[i-1]) / (2 * dx)
    return math.sqrt(1 + dy_dx**2)

# Gauss-Seidel Method
def gauss_seidel(mu_g_over_Tx, sweeps):
    y = initial_guess()
    
    for s in range(sweeps):
        for i in range(1, N - 1):
            rhs = RHS(y, i)
            y[i] = 0.5 * (y[i+1] + y[i-1]) - 0.5 * mu_g_over_Tx * dx**2 * rhs
    
    return y

# Cord length - Eq 20.21
def calculate_length(y, mu_g_over_Tx):
    
    tau_x_over_mu_g = 1.0 / mu_g_over_Tx
    L = (tau_x_over_mu_g / dx) * ( (y[N-1] - y[N-2]) - (y[1] - y[0]) )
    
    return L

# Test a range of mu_g_over_Tx values
length_values = []

mu_g_over_Tx = np.linspace(0.1, 10, 100)


for m in mu_g_over_Tx:
    y = gauss_seidel(m, sweeps) # Solve for this m
    L = calculate_length(y, m) # Calculate cord length
    length_values.append(L)

# Plot L versus mu_g_over_Tx
plt.close()
plt.plot(mu_g_over_Tx, length_values)
plt.xlabel('g/x')
plt.ylabel('Cord Length L')
plt.title('Exercise 20.7a')
plt.show()
```

**Exercise 20.7b**

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Problem setup
xA, yA = 0.0, 0.0
xB, yB = 1.0, 2.0
N = 101
sweeps = 5000
dx = (xB - xA) / (N - 1)

# Initial solution: straight line y = 2x
def initial_guess():
    y = []
    for i in range(N):
        y.append(2.0 * (i * dx))
    return y

# Right side of differential equation
def RHS(y, i):
    dy_dx = (y[i+1] - y[i-1]) / (2 * dx)
    return math.sqrt(1 + dy_dx**2)

# Gauss-Seidel Method
def gauss_seidel(mu_g_over_Tx, sweeps):
    y = initial_guess()
    
    for s in range(sweeps):
        for i in range(1, N - 1):
            rhs = RHS(y, i)
            y[i] = 0.5 * (y[i+1] + y[i-1]) - 0.5 * mu_g_over_Tx * dx**2 * rhs
    
    return y

# Cord length - Eq 20.21
def calculate_length(y, mu_g_over_Tx):
    tau_x_over_mu_g = 1.0 / mu_g_over_Tx
    L = (tau_x_over_mu_g / dx) * ( (y[N-1] - y[N-2]) - (y[1] - y[0]) )
    return L

# Function F(k) = L(k) - L_target
# k = mu_g_over_Tx
# L(k) is the computed cord length
def F(k, L_target):
    y = gauss_seidel(k, sweeps)
    L = calculate_length(y, k)
    return L - L_target

# Bisection method to find k such that F(k) = 0
def bisection(L_target, tolerance, max_iterations=100): # ini - 10
    
    k_L = 0.1
    k_R = 10.0
    
    F_kL = F(k_L, L_target)
    F_kR = F(k_R, L_target)
    
    # Check that F(k_L) and F(k_R) have opposite signs
    if F_kL * F_kR > 0:
        return None
    
    for iteration in range(1, max_iterations + 1):
        # Compute midpoint
        k_M = (k_L + k_R) / 2.0
        F_kM = F(k_M, L_target)
                
        # Check for convergence
        if abs(k_R - k_L) < tolerance:
            return k_M
        
        # Determine which half contains the root
        if F_kL * F_kM < 0: # Root is in [k_L, k_M]
            k_R = k_M
            F_kR = F_kM
        else: # Root is in [k_M, k_R]
            k_L = k_M
            F_kL = F_kM
    
    print(f"Debugging: DID NOT converge within {max_iterations} iterations, add more iterations")
    return (k_L + k_R) / 2.0

L_target = 3.0
k_solution = bisection(L_target, tolerance=1e-6)

y_solution = gauss_seidel(k_solution, sweeps)
L_computed = calculate_length(y_solution, k_solution)
    
print(f"k = g/x = {k_solution}")
    
# Plot the cord shape
xs = [i * dx for i in range(N)]

plt.close()
plt.plot(xs, y_solution, label='shape')
plt.title('Exercise 20.7b')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()    
    
```
