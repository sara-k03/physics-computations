**Exercise 13.2c**

Write a code to approximate $\int_{a}^{b}f(x)dx$ using the left endpoint rule. Test your code using $f(x)= sin(x)$, a = 0 and b = $\pi$/2. What is the exact answer? Compare $I_L$ to the exact answer for N = 2, 4, 8, 16 and 32.

```python
import numpy as np

def left_endpoint_rule( f, a, b, N ):
    delta_x = ( b - a ) / N
    x = np.linspace( a, b, N + 1 )
    total = 0
    for i in range( 1, N + 1 ):
        total += ( f( x[ i - 1 ] ) * delta_x )

    return total

def f( x ):
    return np.sin( x )

N = [ 2, 4, 8, 16, 32 ]
a = 0
b = np.pi / 2
exact = 1

for i in N: 
    approximate = left_endpoint_rule( f, a, b, i )
    error = 1 - approximate
    print( "N = ", i, ", Approximate:", approximate, ", Error:", error  )
```

**Exercise 13.3b**

Write a code to approximate $\int_{a}^{b}f(x)dx$ using the midpoint endpoint rule. Test your code using $f(x)= sin(x)$, a = 0 and b = $\pi$/2. What is the exact answer? Compare $I_M$ to the exact answer for N = 2, 4, 8, 16 and 32. Compare the midpoint rule to the left and right endpoint rules.

```python
import numpy as np

def left_endpoint_rule( f, a, b, N ):
    delta_x = ( b - a ) / N
    x = np.linspace( a, b, N + 1 )
    total = 0
    for i in range( 1, N + 1 ):
        total += ( f( x[ i - 1 ] ) * delta_x )
    return total

def right_endpoint_rule( f, a, b, N ):
    delta_x = ( b - a ) / N
    x = np.linspace( a, b, N + 1 )
    total = 0
    for i in range( 1, N + 1 ):
        total += ( f( x[ i ] ) * delta_x )
    return total

def midpoint_rule( f, a, b, N ):
    delta_x = ( b - a ) / N
    x = np.linspace( a, b, N + 1 )
    total = 0
    for i in range( 1, N + 1 ):
        total += ( f( ( x[ i - 1 ] + x[ i ] ) / 2 ) * delta_x )
    return total

def f( x ):
    return np.sin( x )

N_values = [ 2, 4, 8, 16, 32 ]
a = 0
b = np.pi / 2
exact = 1

print(f"{'N':<5}{'Midpoint':>15}{'Left':>15}{'Right':>15}{'Exact':>15}{'Error(Mid)':>15}")

for N in N_values:
    left = left_endpoint_rule(f, a, b, N)
    right = right_endpoint_rule(f, a, b, N)
    midpoint = midpoint_rule(f, a, b, N)
    err_mid = abs(midpoint - exact)
    
    print(f"{N:<5}{midpoint:15.10f}{left:15.10f}{right:15.10f}{exact:15.10f}     {err_mid:15.10e}")
```

**Exercise 13.4a**

As in exercise 13.2c, use the left endpoint rule to approximate the integral $\int_{0}^{\pi/2}sin(x)dx$. Since the exact answer is known, you can carry out a convergence test: Compute the error at multiple resolutions. Copy and paste your results into a text file with N in the first column and the error in the second column. Create a code to read the text file and plot log|error| versus log(N). Does the curve approach a straight line with slope m = 1 as the resolution is increased?

ANSWER: The slope does appear to approach -1. 

```python
import numpy as np
import matplotlib.pyplot as plt

def left_endpoint_rule( f, a, b, N ):
    delta_x = ( b - a ) / N
    x = np.linspace( a, b, N + 1 )
    total = 0
    for i in range( 1, N + 1 ):
        total += ( f( x[ i - 1 ] ) * delta_x )
    return total

def f( x ):
    return np.sin( x )

a = 0
b = np.pi / 2
exact = 1

N = np.arange( 10, 1001 ) # Resolution of 1 through 1000 
approx = [ ]

for i in N:
    approx.append( left_endpoint_rule( f, a, b, i ) )

approx = np.array( approx )
error = np.abs( exact - approx )

combined = np.array( [ N, error ] ).T
np.savetxt( "output.txt", combined)

# Plotting 
plt.plot( np.log( N ), np.log( error ) )
plt.xlabel( "log(N)" )
plt.ylabel( "log(|error|)" )
plt.title( "Left Endpoint Rule Convergence" )
plt.grid( True )
plt.show()

# Slope over the last 50 points:
slope = np.diff( np.log( error ) ) / np.diff( np.log( N ) )
print("Approximate slope (average over last 50 points):", np.mean( slope[ -50: ] ) )
```

**Exercise 13.4d**

Write a code to compute:

$\int_{1}^{10}xln(10/x)dx$

using the left endpoint and midpoint rules, where ln is the base e logarithm. 

What is the exact answer? 

With the left endpoint rule, approximately how many subintervals N are required to yield the answer with an error less than 0.01? Less than 0.001? 

With the midpoint rule, approximately how many subintervals N are required to yield the answer with an error less than 0.01? Less than 0.001?

ANSWER: (from code output)
    
    Exact: 23.59870745350298 

    Subintervals for error below 0.01 for LEFT ENDPOINT: 1035
    Subintervals for error below 0.001 for LEFT ENDPOINT: 10361 
    
    Subintervals for error below 0.01 for MIDPOINT: 28
    Subintervals for error below 0.001 for MIDPOINT: 89

```python
import sympy as sp
import numpy as np

def left_endpoint_rule( f, a, b, N ):
    delta_x = ( b - a ) / N
    x = np.linspace( a, b, N + 1 )
    total = 0
    for i in range( 1, N + 1 ):
        total += ( f( x[ i - 1 ] ) * delta_x )
    return total

def midpoint_rule( f, a, b, N ):
    delta_x = ( b - a ) / N
    x = np.linspace( a, b, N + 1 )
    total = 0
    for i in range( 1, N + 1 ):
        total += ( f( ( x[ i - 1 ] + x[ i ] ) / 2 ) * delta_x )
    return total

def f( x ):
    # np.log is natural log
    return x * np.log( 10 / x )

# Exact answer
x = sp.symbols( 'x' );
f_sp = x * sp.ln( 10 / x )
exact = float( sp.integrate( f_sp, ( x, 1, 10 ) ) )
print("Exact:", exact, "\n")

# Finding N for error = 0.01, error = 0.0001 for Left Endpoint Rule
N = 1 
a = 1
b = 10

approx = left_endpoint_rule( f, a, b, N )
error = np.abs( exact - approx )

while error > 0.01:
    N += 1
    approx = left_endpoint_rule( f, a, b, N )
    error = np.abs( exact - approx )
print("Subintervals for error below 0.01 for LEFT ENDPOINT:", N)

while error > 0.001:
    N += 1
    approx = left_endpoint_rule( f, a, b, N )
    error = np.abs( exact - approx )
print("Subintervals for error below 0.001 for LEFT ENDPOINT:", N, "\n")

# Finding N for error = 0.01, error = 0.0001 for Midpoint Rule
N = 1 
a = 1
b = 10

approx = left_endpoint_rule( f, a, b, N )
error = np.abs( exact - approx )

while error > 0.01:
    N += 1
    approx = midpoint_rule( f, a, b, N )
    error = np.abs( exact - approx )
print("Subintervals for error below 0.01 for MIDPOINT:", N)

while error > 0.001:
    N += 1
    approx = midpoint_rule( f, a, b, N )
    error = np.abs( exact - approx )
print("Subintervals for error below 0.001 for MIDPOINT:", N)
```

**Exercise 13.5b**

Use the midpoint rule to evaluate the integral

$\int_{-10}^{10}ln(2 + sin(x))dx$

Make a table showing $N$ and $I_M$ for increasing values of $N$. Use your data to determine the value of the integral to 6 significant figures.

```python
def midpoint_rule( f, a, b, N ):
    delta_x = ( b - a ) / N
    x = np.linspace( a, b, N + 1 )
    total = 0
    for i in range( 1, N + 1 ):
        total += ( f( ( x[ i - 1 ] + x[ i ] ) / 2 ) * delta_x )
    return total

def f( x ):
    # np.log is natural log
    return np.log( 2 + np.sin( x ) )

N_vals = np.arange( 1, 201 ) 
a = -10
b = 10

for N in N_vals: 
    approx = midpoint_rule( f, a, b, N )
    print( f"{N} {approx}" )
```
