**Exercise 7.2a**

Write a function that computes the factorial of any positive integer. Use this function in a program that asks the user to input two positive integers, m and n, then computes the number of combinations of n objects taken m at a time:

(n  m) = (n!)/(m!*(n-m)!)

Test your code to make sure it works correctly. 

```python
def factorial(x):
    result = 1 
    while x > 1:
        result = result * x
        x = x - 1
    return result

# n choose m
n = 5
m = 2

result = (factorial(n)) / (factorial(m) * factorial(n - m))
print(result)
```

**Exercise 7.2c**

Create a Python program that defines the function sin(a/x). Have your code produce a graph of sin(3/x) for 0.1 <= x <= 2.0. The inputs to the function will be a = 3 and a NumPy array of x values. The function will return a NumPy array of sin(a/x) values.

```python
import numpy as np
import matplotlib.pyplot as plt

def function(a, x):
    return np.sin(a/x)

a = 3
x = np.linspace(0.1, 2.0, 100) # 100 values in between
f = function(a, x)

plt.close()
plt.plot(x, f) 
plt.title("Graph of sin(3/x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

**Exercise 7.3c**

Newton’s universal law of gravitation says that the gravitational force between objects of mass m1 and m2 is:

F = (G * m1 * m2)/(r ** 2)

where G = 6.67 * 10 ^ -11 N * m^2/ kg^2 is Newton’s constant. Here r is the distance between the two objects. Write a program to compute the force between Earth (m2 = 5.97 * 10^24 kg) and the moon (m1 = 7.35 * 10^22kg) at various distances. Use a function definition for F, treating r as a variable and G, m1 and m2 as parameters. The distance between Earth and the moon varies from roughly r = 3.6 * 10^8 m and r = 4.1 * 10^8 m. How much does the force vary? 

ANSWER: 5.172202485706254e+19 N

```python
import numpy as np

def gravitation(G, m1, m2, r):
    F = (G * m1 * m2)/(r ** 2)
    return F

# Parameters 
G = 6.67 * 10 ** -11
m1 = 7.35 * 10 ** 22
m2 = 5.97 * 10 ** 24 

# Variables
r1 = 3.6 * 10 ** 8
r2 = 4.1 * 10 ** 8

# Difference 
f1 = gravitation(G, m1, m2, r1)
f2 = gravitation(G, m1, m2, r2)

print(f1)
print(f2)

print(f1 - f2)
```

**Exercise 7.5c**

Write a code that accepts a word as user input and counts the number of vowels in the word. (Vowels are the letters “a”, “e”, “i”, “o” and “u”.)

```python
word = input()
counter = 0

for i in range(len(word)):
    if word[i] == 'a' or word[i] == 'e' or word[i] == 'i' or word[i] == 'o' or word[i] == 'u':
        counter = counter + 1

print(counter)
```

**Exercise 7.6c**

Create an array containing all of the numbers of the form 3^n (where n is a positive integer) that are less than 10000.

```python
arr = []

n = 1
num = 3 ** n

while num < 10000:
    arr.append(num)
    n = n + 1
    num = 3 ** n

print(arr)
```

**Exercise 7.7b**

The Chebyshev polynomials of the first kind are defined by:

T_n(x) = cos(n * arccos(x))

where n is a nonnegative integer and -1 <= x <= 1. (You can use multiple–angle trig identities to show that T_n(x) is an nth order polynomial in x.) Create a function definition for the Chebyschev polynomials and plot T0 through T5 on the same graph.

```python
import numpy as np

#n has to 0 or above, x is some value between -1 and 1
def T(n, x): 
    parentheses = n * np.arccos(x)
    return np.cos(parentheses)

x = np.linspace(-1, 1, 100) # 100 values in between
t0 = T(0, x)
t1 = T(1, x)
t2 = T(2, x)
t3 = T(3, x)
t4 = T(4, x)
t5 = T(5, x)

plt.close()

plt.plot(x, t0)
plt.plot(x, t1)
plt.plot(x, t2)
plt.plot(x, t3)
plt.plot(x, t4)
plt.plot(x, t5)

plt.title("T0 - T5 Chebyshev polynomials")
plt.xlabel("x")
plt.ylabel("T(x)")

plt.show()
```
