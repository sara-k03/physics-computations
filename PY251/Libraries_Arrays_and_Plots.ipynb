**Exercise 4.5a**
Create a new version of your projectile code. Use the linspace command to create a time array with 100 or more values. Use a single command (not a for loop) to compute the array of x values. Similarly for the y array. Have your code plot the shape of the trajectory, y versus x.

```python
# Exercise 4.5a

import numpy as np
import matplotlib.pyplot as plt

x0 = 2000 # x-component of initial position (m) (arbitrary)
y0 = 2000  # y-component of initial position (m) (arbitrary)
g = 9.8 # acceleration due to gravity (m/s**2) (constant)
v0x = 5 # x-component of initial velocity (m/s) (arbitrary)
v0y = 55 # y-component of initial velocity (m/s) (arbitrary)

t = np.linspace(1.0, 50.0, 100) # 100 values between 1 and 50 (increments of 1/2)
x = x0 + v0x*t
y = y0 + v0y*t- 0.5*g*t**2

plt.close()
plt.plot(x, y) 
plt.title("Trajectory")
plt.xlabel("X-position")
plt.ylabel("Y-position")
plt.show()
```

**Exercise 4.5b**
Plot the functions f1(x) = x^3 and f2(x) = x^4 from -1 to 1. Use linspace to create an array of x values, then create arrays y1 = x ** 3 and y2 = x ** 4. Plot the two functions on the same graph by using two plt.plot commands. You might want to choose different colors for the two curves.

```python
# Exercise 4.5b

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y1 = x**3
y2 = x**4
plt.plot(x, y1, color="red")
plt.plot(x, y2, color="blue")
plt.title("Function x**3 (red) and Function x**4 (blue)")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
