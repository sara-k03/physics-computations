**Exercise 21.2a**

Write a code to solve the wave equation with boundary conditions $y(t, x_a) = y(t, x_b) = const$ at the endpoints $x_a = -5.0$ and $x_b = 5.0$ (in SI units).  Use the CTCS scheme with J = 100 or more. Choose the mass density Âµ = 0.01 and tension T = 25.0, which yields a
wave speed of c = 50.0. Start the string from rest, Eq. (21.13), with a Gaussian shape, Eq. (21.12). Choose any value for the constant
A, such as A = 1. Choose a value for  such that the endpoint values $y_0^0$ and $y_J^0$ are close to zero. Have your code plot y versus x at various times on the same graph.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters 
xa = -5
xb = 5
mu = 0.01   # mass density
T = 25.0    # tension
c = np.sqrt(T / mu)  # wave speed -> 50.0
J = 100 # number of space intervals 
N = 500 # number of timesteps 

dx = (xb - xa) / J
dt = 0.2 * dx / c 

x = np.linspace(xa, xb, J + 1)

# Allocate arrays
ym = np.zeros(J + 1)  
y  = np.zeros(J + 1)
yp = np.zeros(J + 1)

# Initial conditions 
A = 1
sigma = 1
y0 = A * np.exp(-(x**2) / sigma**2)

y = np.copy(y0)
yp = np.copy(y0)

# Next step 
ym = np.copy(y)
y = np.copy(yp)

# Looping 

# r = (c * dt/dx) ** 2
for n in range(2, N + 1):
    
    for j in range(1, J): 
        yp[j] = 2.0 * y[j] - ym[j] + (c * dt/dx) ** 2 * (y[j + 1] - 2.0 * y[j] + y[j - 1]) 

    # Boundary conditions
    yp[0] = y[0] 
    yp[J] = y[J]
        
    ym = np.copy(y) 
    y = np.copy(yp)
        
    if n % 100 == 0:
        plt.plot(x, y, label=f"t = {n*dt:.2f}")

# Final plot 
plt.title("Exercise 21.2a")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

**Exercise 21.2b**

Estimate the location of the peak in y(t, x) at various times, and use this to verify that the wave speed is c = 50.0.

(Peak coordinates are in x, y form) 

* Estimated peaks at t = 0.04 --> (-2, 0.5) and (2, 0.5)
* Estimated peaks at t = 0.08 --> (-4, 0.5) and (4, 0.5)
* Estimated peaks at t = 0.12 --> (-4, -0.5) and (4, -0.5)
* Estimated peaks at t = 0.16 --> (-2, -0.5) and (2, -0.5)
* Estimated peak at t = 0.2 --> (0, -1)

According to the script below, when peaks on curves on the same side of the y-axis are taken into account, the speed comes out to be c=50. 

```python
import numpy as np

data = [
    (0.04,  2),
    (0.08,  4),
    (0.12,  4),
    (0.16,  2),
    (0.20,  0),
]

speeds = []
for i in range(len(data) - 1):
    t1, x1 = data[i]
    t2, x2 = data[i + 1]
    if t2 != t1:
        c_est = abs(x2 - x1) / abs(t2 - t1)
        speeds.append(c_est)
        print(f"Between t={t1} and t={t2}: speed â {c_est:2f}")
```

**Exercise 21.3b**

Run numerical experiments to find $\eta_{max}$ (approximately) for the CTCS algorithm.

The values of $\eta$ are only stable until 1. Any value above that results in an unstable graph. Even $\eta$ = 1.01 starts to become slightly unstable. 

```python
import numpy as np
import matplotlib.pyplot as plt

values = [0.2, 0.4, 0.6, 0.8, 1, 1.01]
for i in values:
    # Parameters 
    xa = -5
    xb = 5
    mu = 0.01   # mass density
    T = 25.0    # tension
    c = np.sqrt(T / mu)  # wave speed -> 50.0
    J = 100 # number of space intervals 
    N = 500 # number of timesteps 
    
    dx = (xb - xa) / J
    dt = i * dx / c 
    
    x = np.linspace(xa, xb, J + 1)
    
    # Allocate arrays
    ym = np.zeros(J + 1)  
    y  = np.zeros(J + 1)
    yp = np.zeros(J + 1)
    
    # Initial conditions 
    A = 1
    sigma = 1
    y0 = A * np.exp(-(x**2) / sigma**2)
    
    y = np.copy(y0)
    yp = np.copy(y0)
    
    # Next step 
    ym = np.copy(y)
    y = np.copy(yp)
    
    # Looping 
    
    # r = (c * dt/dx) ** 2
    for n in range(2, N + 1):
        
        for j in range(1, J): 
            yp[j] = 2.0 * y[j] - ym[j] + (c * dt/dx) ** 2 * (y[j + 1] - 2.0 * y[j] + y[j - 1]) 
    
        # Boundary conditions
        yp[0] = y[0] 
        yp[J] = y[J]
            
        ym = np.copy(y) 
        y = np.copy(yp)
            
        if n == 100: # Only want it plotted ONCE per time step value
            plt.plot(x, y, label=f"$\eta$ = {i:.2f}")

# Final plot 
plt.title("Exercise 21.3b")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
